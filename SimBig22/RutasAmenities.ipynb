```python
import pandas as pd
import geopandas as gpd
import GOSTnets as gn
import time
import logging
```

```python
logging.basicConfig(level=logging.INFO,
                    format='%(asctime)s :: %(levelname)s :: %(name)s :: %(process)d :: %(message)s',
                    datefmt='%Y-%m-%d %H:%M:%S')
logger = logging.getLogger('MAIN')
```

```python
%%time
grafotransporte_nodes = pd.read_csv("rutas/GrafoTransporte_nodes.csv")
gs = gpd.GeoSeries.from_wkt(grafotransporte_nodes['geometry'])
grafotransporte_nodes = gpd.GeoDataFrame(grafotransporte_nodes, geometry=gs).set_index('osmid')
logger.info("Loading GrafoTransporte_nodes")
grafotransporte_nodes.head()
```

```python
%%time
amenities = pd.read_csv("LimaOSM/LimaAmenities.csv")
gs = gpd.GeoSeries.from_wkt(amenities['geometry'])
amenities = gpd.GeoDataFrame(amenities, geometry=gs).set_index('osmid')
logger.info("Loading LimaAmenities")

amenities.head()
```

```python
hospital = pd.read_csv("LimaOSM/HospitalsOSM.csv")
gs = gpd.GeoSeries.from_wkt(hospital['geometry'])
hospital = gpd.GeoDataFrame(hospital, geometry=gs).set_index('osmid')

hospital['geometry'] = hospital.apply(lambda x: x['geometry'].centroid, axis=1)
hospital.head()
```

```python
# hospital.plot()
```

```python
# amenities_filter = ['bank','casino','clinic','courthouse','hospital','marketplace', 'police','school','university']
AMENITY_NAME = 'hospital'

mask = [AMENITY_NAME]
# amenity = amenities.query('amenity in @mask')
amenity = hospital.query('amenity in @mask')
amenity.head()
```

```python
%%time

start = time.time()
logger.info("Snap process begin")

pandana = gn.pandana_snap_points( grafotransporte_nodes , amenity )

logger.info("Process finished in %.3f seconds", time.time() - start)
```

```python
%%time

amenityid = "{}_osmid".format(AMENITY_NAME)
amenitydist = "{}_dist".format(AMENITY_NAME)
amenityid, amenitydist
pandana.rename(columns={'NN':amenityid, 'NN_dist': amenitydist}, inplace=True)
pandana.head()
```

```python
%%time
filename = "rutas/GrafoTransporteNodes_{}.csv".format(AMENITY_NAME.capitalize())
filename
pandana.to_csv(filename)
```
